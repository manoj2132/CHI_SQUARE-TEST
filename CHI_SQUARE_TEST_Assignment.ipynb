**State the Hypotheses:**

A statistical technique called hypothesis testing is used to determine if a certain claim (hypothesis) about a population is sufficiently supported by the evidence in a sample of data. It helps in determining whether there is a substantial difference between the observed data and what we would anticipate given a particular assumption.



**Compute the Chi-Square Statistic:**

```python
#step1: import libaries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df=pd.read_csv('/content/CHi-square assignment - Sheet1.csv')
df
```

```python
#step:2 some exploratory data analysis
df.info()
```

```python
df.describe()
```

```python
#step 3:we are trying to find the relation ship between smoker and gender
dataset_table=pd.crosstab(df['Smart Thermostat'],df['Smart Light'])
dataset_table
```

```python
#step 4 : observed value
observed_values=dataset_table.values
observed_values
```

```python
#step:5=import chiz contugency table
import scipy.stats as stats
import numpy as np
from  scipy.stats import chi2_contingency
val=stats.chi2_contingency(dataset_table)
val
#expected frequency =(row total*column total)/n
```

```python
expected_values=val[3]
expected_values
```

```python
no_of_rows=len(dataset_table.iloc[0:2,0])
no_of_columns=len(dataset_table.iloc[0,0:2])
ddof=(no_of_rows-1)*(no_of_columns-1)
alpha=0.05
ddof
```

```python
#formula for chi2 (xi=(o-e)^2/e))
#(x=((o-e)**2)/e)
#** indicates exponetial to the power of 2
from scipy.stats import chi2
chi_square=sum([(o-e)**2./e for o,e in zip(observed_values,expected_values)])
chi_square
# observed value
#expected value
```

```python
chi_square_statistic=chi_square[0]+chi_square[1]
#2*2 matrix,row(0) and row(1)
chi_square_statistic
```

```python
#percent point function(inverse of cdf)
critical_value=chi2.ppf(q=1-alpha,df=ddof)
critical_value
```

```python
p_value=1-chi2.cdf(x=chi_square_statistic,df=ddof)
p_value
#cumulative distribution function
```

**Make a Decision:**

```python
if chi_square_statistic>=critical_value:
  print("reject H0 ,There is a relationship between 2 categorical variables")
else:
   print("reject H0 ,There is a  no relationship between 2 categorical variables")
```

```python
if p_value<=critical_value:
   print("reject H0 ,There is a relationship between 2 categorical variables")
else:
   print("reject H0 ,There is a  no relationship between 2 categorical variables")
```
